##1.Immobili
Definire la classe Immobile con i seguenti attributi

*   codice
*   metri quadrati
*   tipologia che può assumere solo i valori : monolocale, bilocale, trilocale, quadrilocale
*   indirizzo
*   città
*   prezzo
*   venduto (inizialmente uguale a false)


Crea un oggetto, assegna dei valori e visualizzali utilizzando il metodo print(oggetto)


```python
class Immobile:
  def __init__(self,codice, metriquadri, tipologia, indirizzo, citta, prezzo):
    self.codice=codice
    self.metriquadri=metriquadri
    self.tipologia=tipologia
    self.indirizzo=indirizzo
    self.citta=citta
    self.prezzo=prezzo
    self.venduto=False

  def __str__(self):
    vis = ""
    vis += f"{self.codice}; {self.metriquadri} mq; {self.tipologia}; {self.indirizzo}; {self.citta}; prezzo: {self.prezzo} euro; Venduto: {self.venduto}"
    return vis
```

```python

#I modalità
immobile1=Immobile("123",100,"bilocale","via Napoli,4","Magenta",1200)
print(immobile1)
#output: codice:123, mq:100, tipologia:bilocale, indirizzo:Via Napoli 4, citta:Magenta, prezzo:1200€ venduto:False

codice=input("inserisci il codice dell'immobile: ")
mq=int(input("inserisci i mq (es. 90 :"))
tipo=input("inserisci la tipologia: ")
indirizzo=input("inserisci l'indirizzo: ")
citta=input("inserisci la città: ")
prezzo=int(input("inserisci il prezzo: "))
#non chiedo in input il valore da assegnare all'attributo venduto perchè di default assume il valore False
immobile2=Immobile(codice,mq,tipo,indirizzo,citta,prezzo)
print(immobile2)
```

##3.Immobili

Crea una classe Agenzia che preveda i seguenti metodi:

1.   Aggiunta di un immobile (utilizza valori di test tutti uguali senza inserire ogni volta i valori a mano)
2.   Vendita di un immobile
3.   Aggiorna prezzo
4.   Visualizza immobili appartenenti ad una tipologia
5.   Cerca un immobile
6.   Visualizza tutti gli immobili
7.   Visualizza il totale dei prezzi di tutti gli immobili venduti
8.   Visualizza il prezzo medio di una tipologia fornita in input






```python
class Agenzia:
  def __init__(self):
    self.listaImmobili = []

  def aggiungiImmobile(self, imm):
    for i in range(len(self.listaImmobili)):
      if self.listaImmobili[i].codice == imm.codice:
        print("Immobile già presente nella lista")
        return False
    self.listaImmobili.append(imm)

  def vendiImmobile(self, codice):
    for imm in self.listaImmobili:
      if imm.codice == codice:
        imm.venduto = True
        print("venduto")
        break

  def aggiornaPrezzo(self, codice, nuovoPrezzo):
    for imm in self.listaImmobili:
      if imm.codice == codice:
        imm.prezzo = nuovoPrezzo
        break

  def visualizzaImmobiliTipo(self, tipologia):
    for imm in self.listaImmobili:
      if imm.tipologia == tipologia:
        print(imm)

  def ricerca(self,codice):
    for imm in self.listaImmobili:
      if imm.codice == codice:
        return imm
    return None

  def visualizzaTutto(self):
    for imm in self.listaImmobili:
      print(imm)

  def totPrezziVenduti(self):
    tot = 0
    for imm in self.listaImmobili:
      if imm.venduto:
        print(imm)
        tot += imm.prezzo
    return tot

  def prezzoMedioTipo(self, tipologia):
    tot = 0
    cont = 0
    for imm in self.listaImmobili:
      if imm.tipologia == tipologia:
        tot += imm.prezzo
        cont += 1
    if cont == 0:
      return 0
    else:
      return tot / cont

if __name__ == "__main__":
  agenzia = Agenzia()
  metriquadri=0;tipologia="";prezzo=0;nuovoPrezzo=0;

  codice=input("inserisci il codice dell'immobile: ")
  while metriquadri<=0:
    metriquadri=int(input("inserisci i metri quadri dell'immobile :"))
  while (tipologia !="monolocale" and tipologia !="bilocale"and tipologia !="trilocale" and tipologia !="quadrilocale"):
    tipologia=input("inserisci la tipologia: ")
  indirizzo=input("inserisci l'indirizzo: ")
  citta=input("inserisci la città: ")
  while prezzo<=0:
    prezzo=int(input("inserisci il prezzo: "))
  immobile=Immobile(codice,metriquadri,tipologia,indirizzo,citta,prezzo)
  agenzia.aggiungiImmobile(immobile);

  agenzia.aggiungiImmobile (Immobile("123",100,"bilocale","via Milano,97","Milano",1200))
  agenzia.aggiungiImmobile (Immobile("987",75,"bilocale","via Roma,45","Magenta",1000))
  codice=input("inserisci il codice dell'immobile da vendere: ")
  agenzia.vendiImmobile(codice)
  codice=input("inserisci il codice dell'immobile da cambiare il prezzo: ")
  while nuovoPrezzo<=0:
    nuovoPrezzo=int(input("inserisci il nuovo prezzo dell'immobile: "))
  agenzia.aggiornaPrezzo(codice, nuovoPrezzo)
  while (tipologia !="monolocale" and tipologia !="bilocale"and tipologia !="trilocale" and tipologia !="quadrilocale"):
    tipologia=input("inserisci la tipologia dell'immobile(monolocale, bilocale, trilocale, quadrilocale): ")
  agenzia.visualizzaImmobiliTipo(tipologia)
  codice=input("inserisci il codice dell'immobile da cercare: ")
  if(agenzia.ricerca(codice)==None):
    print("Immobile non trovato")
  else:
    print(agenzia.ricerca(codice))
  agenzia.visualizzaTutto()
  print(agenzia.totPrezziVenduti())
  tipologia=""
  while (tipologia !="monolocale" and tipologia !="bilocale"and tipologia !="trilocale" and tipologia !="quadrilocale"):
    tipologia=input("inserisci la tipologia dell'immobile(monolocale, bilocale, trilocale, quadrilocale): ")
  print(agenzia.prezzoMedioTipo(tipologia))
```
